# Analysis of Agent Performance

## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_theme()
```

## Graphs

```python
data = pd.read_csv("./client/rl_out/progress.csv")

plt.figure()
ax = sns.lineplot(data=data, x="time/iterations", y="rollout/ep_rew_mean")
ax.set(xlabel="Iteration", ylabel="Mean Episode Reward")

plt.figure()
ax = sns.lineplot(data=data, x="time/iterations", y="rollout/ep_len_mean")
ax.set(xlabel="Iteration", ylabel="Mean Episode Length")

plt.figure()
ax = sns.lineplot(data=data, x="time/iterations", y="train/loss")
ax.set(xlabel="Iteration", ylabel="Loss")
```
